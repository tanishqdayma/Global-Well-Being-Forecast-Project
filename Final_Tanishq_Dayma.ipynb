### Definition of variables:

**Primary columns:**<br>
**1. Country name:** The name of the country under consideration. Each data row represents a different country.

**2. Ladder score:** It measures a country's level of happiness or quality of life. A higher value suggests a higher level of fulfillment.

**3. Standard error of ladder score:** This is a measurement of the standard error for the "Ladder score." It contributes to the accuracy of the outcomes.

**4. upperwhisker:** This is the upper bound of the confidence interval for the "Ladder score."

**5. lowerwhisker:** This is the lower bound of the confidence interval for the "Ladder score."

**6. Logged GDP per capita:** It reflects the logged Gross Domestic Product (GDP) per capita, a measure of the country's economic well-being.

**7. Social support:** It denotes the degree of social support or social aid available to people of the country.

**8. Healthy life expectancy:** It indicates the predicted length of a healthy life, showing how many years people in that nation may expect to live in good health.

**9. Freedom to make life choices:** It assesses the degree of freedom people have in making life decisions in their respective countries.

**10. Generosity:** It is a measure of generosity or people's eagerness to help others.

**11. Perceptions of corruption:**  It examines people's perceptions of the country's level of corruption. A higher rating might suggest a stronger impression of corruption.

**Additional Columns**
**12. Ladder score in Dystopia:** This is a hypothetical degree of happiness or quality of life in severely adverse conditions known as "Dystopia."

**13. Explained by: Log GDP per capita:** The portion of the "Ladder score" that is explained by the logarithm of GDP per capita.

**14. Explained by: Social support:** It determines how much of the "Ladder score" is explained by social support.

**15. Explained by: Healthy life expectancy:** Describes how the projected healthy life expectancy impacts the "Ladder score."

**16. Explained by: Freedom to make life choices:** Explains the importance of freedom in decision-making in connection to the "Ladder score."

**17. Explained by: Generosity:** It determines how much of the "Ladder score" is explained by generosity.

**18. Explained by: Perceptions of corruption:** This explains how perceptions of corruption influence the "Ladder score."

**19. Dystopia + residual:** This is the total of the values from the "Ladder score in Dystopia" column and the remaining residual impacts that have not been accounted for in other explanations (such as climate, culture, history, and many other personal factors for the country).

## 3. Methodology

### Analytical Methods
In this project, various statistical and modeling techniques were used to understand and predict the World Happiness Scores for different countries in 2023.

1. **Data Preprocessing:**
   - Standard Scaling: Applied Standard Scaler to normalize selected columns, such as GDPperCapita, SocialSupport, LifeExpectancy, FreedomToMakeLifeChoices, Generosity, and Corruption. It transforms the data to have a mean of 0 and a standard deviation of 1, ensuring that all features contribute equally to the analysis.
      
       Why - It was required because regression models are sensitive to the scale of input features. The unit values of Life Expectancy is much higher than other variables and would have dominated the model fitting process due to its larger scales.
   
   - Train-Test Split: Split the dataset into training (80%) and testing (20%) sets.
   
       Why - By dividing our World Happiness Report data into two sets—one for training the model and the other for testing—we can assess how effectively the model predicts happiness scores for countries it hasn't seen before. This ensures that our model generalizes well to new data, enhancing its reliability.

2. **Regression Models:**
   - Linear Regression: Utilized linear regression as a baseline model to establish a simple relationship between the target variable - Happiness Score and all the other variables.

       Why - It helps in the interpretation of the relationship between the features and the target variable which can act as a benchmark for all the other models.
   
   - Lasso Regression: Applied Lasso regression for feature selection and regularization by adding a penalty term to the coefficients based on their impact on the target variable.
   
       Why - Lasso helps in selecting the most relevant features by setting the coefficients of less important features to zero. This aids in simplifying the model and reducing the risk of overfitting.
   
   - Ridge Regression: Ridge regression is used when we want to mitigate issues like mu multicollinearity when independent variables are highly correlated.
   
       Why - This is important for our analysis to ensure that the relationships between various factors influencing happiness scores are accurately captured without being overly influenced by correlations among those factors.
   
   - Elastic Net Regression: Used Elastic Net regression as it combines the strengths of both Lasso and Ridge regression techniques. 
   
       Why -  In our analysis, it allows us to simultaneously address multicollinearity concerns and identify the most influential factors contributing to happiness scores, offering a comprehensive approach to model regularization.
   
   - Polynomial Regression: It is used as it helps capture non-linear relationships between independent and dependent variables.
   
       Why - In instances where the association between features (such as GDP per capita, social support, life expectancy, etc.) and the happiness score is not strictly linear, polynomial regression enables the model to account for more intricate patterns.

### Model Evaluation and Selection

#### Choice of Features or Feature Selection Strategy

In crafting our feature set (X variable), we've omitted 'Country' and 'HappinessScore.' The rationale behind excluding 'Country' is rooted in its categorical nature, unique to each row, making it unsuitable for direct quantitative analysis. As for 'HappinessScore,' it is our target variable.

In contrast, the 'y' variable, representing our target variable, exclusively centers around the 'HappinessScore' which is the key metric of interest.

#### Evaluation Metrics
- **Mean Squared Error (MSE):** Calculated the MSE for each model to measure the average squared difference between predicted and actual values. Lower MSE values indicate better model performance. However, since MSE is not in the same scale as the original data, it might be less interpretable compared to RMSE.
- **Root Mean Squared Error (RMSE):** Computed the RMSE as the square root of the average of the squared differences between predicted and actual values. It provides a single, easy-to-interpret measure of how well the model is performing, with lower RMSE values indicating better accuracy.
- **R-squared (R2):** Evaluated the R2 score to assess the proportion of variance explained by the models. In general, a higher R2 is desirable, as it suggests that the model is better at explaining and predicting the observed outcomes.

```
# Import necessary libraries
import pandas as pd   # For data manipulation
import numpy as np    # For numerical computations

# Visualization libraries
import matplotlib.pyplot as plt  
import seaborn as sns  


# Load the World Happiness Report Dataset from the CSV for the year 2023
#2 Import dataset
data_2023 = pd.read_excel('DataForFigure2.1WHR2023.xls')

data_2023.head()

data_2023.shape

# 2. Check types of data by .info() ; .dtype
data_2023.info()


# 3. For quantitative variable, generate a table for the count, mean, standard deviation, minimum and maximum values and the quantities of the data 
data_2023.describe()


# Generate Sub-dataset
# 1. Make sub-dataset from the original dataset for the research objective, goals by dropping unnecessary variables 
happiness_data_2023 = data_2023.drop(columns=['upperwhisker',
                                    'lowerwhisker',
                                    'Standard error of ladder score',
                                    'Ladder score in Dystopia',
                                    'Explained by: Log GDP per capita', 
                                    'Explained by: Social support', 
                                    'Explained by: Healthy life expectancy',
                                    'Explained by: Freedom to make life choices',
                                    'Explained by: Generosity',
                                    'Explained by: Perceptions of corruption',
                                    'Dystopia + residual'])
happiness_data_2023.head()


# 2. Change the variable names
happiness_data_2023 = happiness_data_2023.rename(columns={
    'Country name': 'Country',
    'Ladder score': 'HappinessScore',
    'Logged GDP per capita': 'GDPperCapita',
    'Social support': 'SocialSupport',
    'Healthy life expectancy': 'LifeExpectancy',
    'Freedom to make life choices': 'FreedomToMakeLifeChoices',
    'Perceptions of corruption': 'Corruption'
})

happiness_data_2023.head()

# 3. check if there are missing values
happiness_data_2023.isnull().sum()


# since the null values identified is only present for one of the column for one of the row. Hence we will be filling it as the median of the rest of the rows for that column.
happiness_data_2023['LifeExpectancy'].fillna(happiness_data_2023['LifeExpectancy'].median(), inplace=True)

# 4. Check if there are duplicated values
happiness_data_2023.duplicated().sum()


# Box Plot Analysis

# Generate Box Plots to check Outliers

for col in happiness_data_2023.columns:
    dtype = happiness_data_2023.dtypes[col]
    
    if dtype == 'object':
        pass
    else:
        sns.boxplot(x=happiness_data_2023[col])
        plt.title(f'Boxplot of {col}')
        plt.show()
        

from scipy import stats

def outliers_z(value):
    # Calculate z-scores for each data point
    z_scores = stats.zscore(value)

    # Define a threshold (e.g., 3 standard deviations)
    threshold = 3.5

    # Identify outliers
    outliers = (z_scores.abs() > threshold)
    
    count_outliers = outliers.sum()
    return count_outliers

for col in happiness_data_2023.columns:
    
    dtype = happiness_data_2023.dtypes[col]
    
    if dtype == 'object':
        print(f"{col} is a qualitative variable")
    else:  # Check for quantitative columns
        outliers = outliers_z(happiness_data_2023[col])
        print(f"{col} has {outliers} outliers")

### EDA
# 1. Generate Histogram to understand the distribution of the data

happiness_data_2023.hist(bins=50, figsize=(20,15))
plt.show()

# PLotting the corelation matrix

# 2. Compute pairwise correlation of columns
correlation_matrix = happiness_data_2023.corr().round(2)

# Use seaborn heatmap to visualize correlations on matrix 
plt.figure(figsize=(12,9))
sns.heatmap(data=correlation_matrix , annot=True)

# calculate score:

correlation = happiness_data_2023.corr('pearson')
correlation['HappinessScore'].sort_values(ascending=False)


# 3. Scatter Plot with the regression line

import seaborn as sns
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'happiness_data_2023' with the specified columns
columns = ["GDPperCapita", 'SocialSupport', 'LifeExpectancy', 'FreedomToMakeLifeChoices', 'Generosity', 'Corruption']

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))

# Flatten the 2D array of subplots for easier iteration
axes = axes.flatten()

# Iterate through each column and create a scatter plot with a fit line
for i, column in enumerate(columns):
    sns.scatterplot(data=happiness_data_2023, x=column, y='HappinessScore', ax=axes[i], color='skyblue', alpha=0.7)
    axes[i].set_title(f'Scatter Plot: {column} vs Happiness Score')

    # Add regression line
    sns.regplot(data=happiness_data_2023, x=column, y='HappinessScore', ax=axes[i], scatter=False, color='orange')

# Adjust the layout
plt.tight_layout()

# Show the plots
plt.show()

#### MODEL TRAINING

# Spliting the data set into independent and dependent variables
X = happiness_data_2023.drop(['Country','HappinessScore'], axis=1)  # Independent variables
y = happiness_data_2023['HappinessScore']    #dependent variables


# Normalize the data using Standard Scaler

from sklearn.preprocessing import StandardScaler

# Assuming you have a DataFrame named 'happiness_data_2023' with the relevant columns
columns_to_normalize = ['GDPperCapita', 'SocialSupport', 'LifeExpectancy', 'FreedomToMakeLifeChoices', 'Generosity', 'Corruption']

scaler = StandardScaler()

# Normalize the selected columns
X[columns_to_normalize] = scaler.fit_transform(X[columns_to_normalize])

# Display the normalized data
print(X.head())

#Train Test split
from sklearn.model_selection import train_test_split

# Splitting the dataset into train(80%) and test(20%) 
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state=40)


# Training the models for Linear, LASSO, Ridge, Elastic Net, and Polynomial Refression

from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score

# Initialize the models
linear_model = LinearRegression()
lasso_model = Lasso()
ridge_model = Ridge()
elasticnet_model = ElasticNet()

# Polynomial Regression
poly = PolynomialFeatures(degree=2)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

# Fit the models on training data 
linear_model.fit(X_train,y_train)
lasso_model.fit(X_train,y_train)
ridge_model.fit(X_train,y_train)
elasticnet_model.fit(X_train,y_train)
poly_model = LinearRegression().fit(X_train_poly, y_train)

# Make predictions on test data 
linear_pred = linear_model.predict(X_test)
lasso_pred = lasso_model.predict(X_test)
ridge_pred = ridge_model.predict(X_test)
elasticnet_pred = elasticnet_model.predict(X_test)
poly_pred = poly_model.predict(X_test_poly)


### MODEL EVALUAION
 
# Calculate MSE and root mean squared error (RMSE) for each model's predictions 
linear_mse = mean_squared_error(y_test, linear_pred)
linear_rmse = np.sqrt(linear_mse)
lasso_mse = mean_squared_error(y_test, lasso_pred)
lasso_rmse = np.sqrt(lasso_mse)
ridge_mse = mean_squared_error(y_test, ridge_pred)
ridge_rmse = np.sqrt(ridge_mse)
elasticnet_mse=mean_squared_error(y_test, elasticnet_pred)
elasticnet_rmse = np.sqrt(elasticnet_mse)
poly_mse = mean_squared_error(y_test, poly_pred)
poly_rmse = np.sqrt(poly_mse)

# Calculate r square for each model's predictions 
linear_r2 = r2_score(y_test, linear_pred)
lasso_r2 = r2_score(y_test, lasso_pred)
ridge_r2 = r2_score(y_test, ridge_pred)
elasticnet_r2 = r2_score(y_test, elasticnet_pred)
poly_r2 = r2_score(y_test, poly_pred)

print(f"Linear MSE: {linear_mse}")
print(f"LASSO MSE: {lasso_mse}")
print(f"Ridge MSE: {ridge_mse}")
print(f"Elastic Net MSE: {elasticnet_mse}")
print(f"Polynomial MSE: {poly_mse}")

print("\n RMSE values:")
print(f"Linear RMSE: {linear_rmse}")
print(f"LASSO RMSE: {lasso_rmse}")
print(f"Ridge RMSE: {ridge_rmse}")
print(f"Elastic Net RMSE: {elasticnet_rmse}")
print(f"Polynomial RMSE: {poly_rmse}")

print("\n R2 values:")
print(f"Linear R2: {linear_r2}")
print(f"LASSO R2: {lasso_r2}")
print(f"Ridge R2: {ridge_r2}")
print(f"Elastic Net R2: {elasticnet_r2}")
print(f"Polynomial R2: {poly_r2}")


### MODEL IMPROVEMENT

# 1. For LASSO Regression
# Applying Cross Validation to Lasso Regression
from sklearn.linear_model import LassoCV

alphas = np.logspace(-4, 4, 200)

lasso_cv = LassoCV(alphas=alphas, cv=5)

# Fit the model
lasso_cv.fit(X_train,y_train)

print(f"Best Alpha after Cross Validation {lasso_cv.alpha_}")

# 2. For Elastic Net Regression
# Applying Grid Search to Elastic Net Regression

from sklearn.model_selection import GridSearchCV
from sklearn.datasets import make_regression

ElasticmodelTest = ElasticNet()
# Define the grid of hyperparameters 'param_grid'
param_grid = {
    'alpha': np.arange(0.01, 1.01, 0.05),
    'l1_ratio': np.arange(0.01, 1.01, 0.05)
}

gridmodel_result = GridSearchCV(estimator=ElasticmodelTest,
                                 param_grid=param_grid,
                                 cv=10).fit(X_train,y_train)

print(f"Best: {gridmodel_result.best_score_} using {gridmodel_result.best_params_}")


## Implementing the model training after getting the values of alphas

from sklearn.linear_model import Lasso, ElasticNet
from sklearn.metrics import mean_squared_error, r2_score

# Initialize the models
lasso_model = Lasso(alpha=0.0001)
elasticnet_model = ElasticNet(alpha=0.21000000000000002)

# Fit the models on training data
lasso_model.fit(X_train,y_train)
elasticnet_model.fit(X_train,y_train)

# Make predictions on test data 
lasso_pred = lasso_model.predict(X_test)
elasticnet_pred = elasticnet_model.predict(X_test)


# Calculate root mean squared error for each model's predictions 

lasso_mse = mean_squared_error(y_test, lasso_pred)
lasso_rmse = np.sqrt(lasso_mse)

elasticnet_mse=mean_squared_error(y_test, elasticnet_pred)
elasticnet_rmse = np.sqrt(elasticnet_mse)

# Calculate r square for each model's predictions 
lasso_r2 = r2_score(y_test, lasso_pred)
elasticnet_r2 = r2_score(y_test, elasticnet_pred)

print(f"LASSO MSE with Hyperparameter Tuning: {lasso_mse}")
print(f"Elastic Net MSE with Hyperparameter Tuning: {elasticnet_mse}")

print("\n RMSE values:")
print(f"LASSO RMSE with Hyperparameter Tuning: {lasso_rmse}")
print(f"Elastic Net RMSE with Hyperparameter Tuning: {elasticnet_rmse}")

print("\n R2 values:")
print(f"LASSO R2 with Hyperparameter Tuning: {lasso_r2}")
print(f"Elastic Net R2 with Hyperparameter Tuning: {elasticnet_r2}")


## Best Model - Elastic Net

# Histogram between the diffence of actual vs predicted values against the frequency
import seaborn as sns
sns.histplot((y_test - linear_pred))

#coefficients

cdf = pd.DataFrame(data = elasticnet_model.coef_, index = X.columns, columns = ['Coefficients'])
print(cdf)

# MODEL PREDICTION:

#Model Prediction

# Example prediction for new data with scaled values
new_data = pd.DataFrame({'GDPperCapita': [(10.228-happiness_data_2023['GDPperCapita'].mean())/happiness_data_2023['GDPperCapita'].std()],
                         'SocialSupport': [(0.842-happiness_data_2023['SocialSupport'].mean())/happiness_data_2023['SocialSupport'].std()],
                         'LifeExpectancy': [(65.700-happiness_data_2023['LifeExpectancy'].mean())/happiness_data_2023['LifeExpectancy'].std()],
                         'FreedomToMakeLifeChoices': [(0.916-happiness_data_2023['FreedomToMakeLifeChoices'].mean())/happiness_data_2023['FreedomToMakeLifeChoices'].std()],
                         'Generosity': [(0.121-happiness_data_2023['Generosity'].mean())/happiness_data_2023['Generosity'].std()],
                         'Corruption': [(0.782-happiness_data_2023['Corruption'].mean())/happiness_data_2023['Corruption'].std()]
                        })

# Predict the happiness score for the new data
predicted_happiness_score = elasticnet_model.predict(new_data)
print(f'Predicted Happiness Score through Elastic Net Model: {predicted_happiness_score[0]}')

```
